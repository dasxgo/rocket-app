## Data according to plan



```python
def calculate_production_values(df_flatfile, well_planned):
    production_columns = ['LiquidsProd_BBL', 'GasProd_MCF', 'WaterProd_BBL']
    production_data = {}

    for column in production_columns:
        data = df_flatfile[column] * (well_planned / df_flatfile['LateralLength_FT']) * (30.4 / df_flatfile['ProducingDays'])
        production_data[column] = data

    df_flatfile[['OIL_BBL', 'GAS_MCF', 'WATER_BBL']] = pd.DataFrame(production_data)

    # Convert columns to int64
    for column in production_columns:
        df_flatfile[['OIL_BBL', 'GAS_MCF', 'WATER_BBL']] = np.array(df_flatfile[['OIL_BBL', 'GAS_MCF', 'WATER_BBL']], np.int64)

    return df_flatfile

well_planned = 5000
df_flatfile= calculate_production_values(df_flatfile, well_planned)
df_flatfile.head()
```
